```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#read data
data = pd.read_csv('WHO-COVID-19-cases.csv')
data.head()
```

```python
# membuat salinan unutuk menjaga keaslian data awal
df = data.copy()
df.head()
```

```python
len(df)
```

```python
# cek status buat tau ada nilai yang hilang
df.isna().sum()
```

```python
# hapus data pake dropna
df.dropna(inplace=True)
```

```python
# cek status lagi
df.isna().sum()
```

```python
len(df)
```

```python
# cek data ada yang terduplicate atau tidak

df.duplicated().sum()
```

```python
# hapus data yang terduplicate

df = df.drop_duplicates()
df.head()
```
